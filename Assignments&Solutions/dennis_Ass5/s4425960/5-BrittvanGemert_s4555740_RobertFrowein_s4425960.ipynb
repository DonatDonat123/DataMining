Hint: To prove the statement rewrite the confidence and the support of the rule in terms of probabilities:

4. If $\frac{p(a,b)}{p(a)}$ < p(b)  
1) Then $\frac{p(\text{~}a,b)}{p(\text{~}a)}$ > $\frac{p(a,b)}{p(a)}$
  
> __Proof__:
  
> $\frac{p(\text{~}a,b)}{p(\text{~}a)}$ > $\frac{p(a,b)}{p(a)}$  
  
> $\frac{p(\text{~}a,b)}{p(\text{~}a)}$ > p(b|a) $\hspace{10mm}$ *(Bayes rule)* 
  
> p($\text{~}$a,b) > p(b|a) $*$ p(~a)  
  
> p($\text{~}$a,b) + p(a,b) > p(b|a) $*$ p($\text{~}$a) + p(a,b) 
  
> p(b) > p(b|a) $*$ p($\text{~}$a) + p(a,b)  
  
> p(b) > $\frac{p(a,b) * p(\text{~}a)}{p(a)}$ + p(a,b)  
  
> p(b) > $\frac{p(a,b) * (1 - p(a))}{p(a)}$ + p(a,b)  
  
> p(b) > $\frac{p(a,b) - p(a,b) * p(a)}{p(a)}$ + p(a,b)  
  
> p(b) > $\frac{p(a,b)}{p(a)} - \frac{p(a,b) * p(a)}{p(a)}$ + p(a,b)  
  
> p(b) > $\frac{p(a,b)}{p(a)}$ - p(a,b) + p(a,b)  
  
> p(b) > $\frac{p(a,b)}{p(a)}$  
  
> $\frac{p(a,b)}{p(a)}$ < p(b)  
  
> This is correct, because it is exactly the same as the if-statement.


2) also $\frac{p(\text{~}a,b)}{p(\text{~}a)}$ > p(b)
  
> __Proof__:
  
> $\frac{p(\text{~}a,b)}{p(\text{~}a)}$ > p(b) 
  
> $\frac{p(\text{~}a,b)}{p(\text{~}a)}$ > p($\text{~}$a,b) + p(a,b)  
  
> p($\text{~}$a,b) > (p($\text{~}$a,b) + p(a,b)) $*$ p($\text{~}$a)  
  
> p($\text{~}$a,b) > (p($\text{~}$a,b) + p(a,b)) $*$ (1 - p(a))  
  
> p($\text{~}$a,b) > p($\text{~}$a,b) - (p($\text{~}$a,b) $*$ p(a)) + p(a,b) - (p(a,b) $*$ p(a))  
  
> $\frac{p(\text{~}a,b)}{p(a)}$ > $\frac{p(\text{~}a,b)}{p(a)}$ - p($\text{~}$a,b) + $\frac{p(a,b)}{p(a)}$ - p(a,b)  
  
> $\frac{p(\text{~}a,b)}{p(a)}$ > $\frac{p(\text{~}a,b)}{p(a)}$ + $\frac{p(a,b)}{p(a)}$ - p(b)  
  
> 0 > $\frac{p(a,b)}{p(a)}$ - p(b)  
  
> This is correct, because in the if-statement p(b) is larger than $\frac{p(a,b)}{p(a)}$, which makes this correct
