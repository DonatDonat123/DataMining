 Suppose we have market basket data consisting of 100 transactions and 20 items. The support for item $ \text{a} = 45 \%$, the support for item $ \text{b} = 80 \%$ and the support for itemset $ \text{ {a,b }} = 30 \%$. Let the support and confidence thresholds be 20$ \%$ and 60$ \%$, respectively.
  
1. (0.5 points) Compute the confidence of the association rule $ \text{ {a } } \rightarrow   \text{{b }} $. Is the rule interesting according to the confidence measure?

2. (0.5 points) Compute the interest measure (or lift, see slide 44 of chapter 6) for the association pattern $ \text{ {a,b}}$. Describe the nature of the relationship between item $ \text{a}$ and item $ \text{b}$  in terms of the itemset measure.

4. (1.5 points) Prove that if the confidence of the rule $ \text{ {a } } \rightarrow   \text{{b }} $ is less than the support of $ \text{ {b }}$  then
$$
c(  \text{ {~a } } \rightarrow   \text{{b }})> c(  \text{ {a } } \rightarrow   \text{{b }})
$$
and
$$
c(  \text{ {~a } } \rightarrow   \text{{b }})> s( {b })
$$

where $ \text{c(.)}$  denotes the rule confidence and  $ \text{s(.)}$ denote the support of an itemset. 

Hint: To prove the statement rewrite the confidence and the support of the rule in terms of probabilities: 
$$
c(  \text{ {a } } \rightarrow   \text{{b }})=  \frac{P( {a,b })}{P( {a })}
$$

and

$$
s( {b })=P( {b })
$$

Further more, make use of probability rules such as $p(b) = p(\text{~}a,b) + p(a,b)$ and $p(\text{~}a) = 1 - p(a)$ and algebraic rules such as $$ \frac{p(a,b)}{p(b)} < p(a) \rightarrow 1 - \frac{p(a,b)}{p(b)} > 1 - p(a) $$


*Double click to type your answer for 5.3.1 here:*

### Answer to question 5.3.1
||$b$|$\neg b$||
|-|-|-|
|$a$|30|15|45|
|$\neg a$|50|5|55|
||80|20|100

*The table above is used to obtain the probabilties for the computations.*

1. $$P(b\ |\ a)=\frac{P(a \cap b)}{P(a)}=\frac{0.30}{0.45}=0.67$$
Confidence of $\{a\} \rightarrow \{b\}$ is $0.67*100\%=67\%$  
Yes, according to the confidence measure this rule is interesting, because it is higher than the confidence threshold ($67\%>60\%$).
2. $$\text{Lift}=\frac{P(b\ |\ a)}{P(b)}=\frac{0.67}{0.80}=0.83$$
Lift of $\{a\} \rightarrow \{b\}$ is 0.83.  
Item $a$ and item $b$ have are negatively associated according to the Lift measure, because the Lift is smaller than 1 ($0.83 < 1$).
3. Although the confidence measure suggests that the rule $\{a\} \rightarrow \{b\}$ is interesting, the Lift indicates that item $a$ and $b$ are negatively associated. This means that the rule $\{a\} \rightarrow \{b\}$ may actually not be that interesting as item $b$ already appears in 80% of the transactions anyway(supp=80%), it is likely that item $b$ will also appear in a number of transactions where item $a$ is present (without necessarily being associated to item $a$).
4. $$c(\{\neg a\} \rightarrow \{b\})=\frac{P(\neg a \cap b)}{P(\neg a)}$$  
$$P(\neg a \cap b)=P(b)-P(a \cap b)=0.8-0.3=0.5$$  
$${P(\neg a)}=1-{P(a)}=1-0.45=0.55$$  
$$c(\{\neg a\} \rightarrow \{b\})=\frac{P(\neg a \cap b)}{P(\neg a)}=\frac{0.5}{0.55}=0.91$$  
If $$c(\{a\} \rightarrow \{b\})=0.67 < s(b)=0.8$$  
then,  
$$c(\{\neg a\} \rightarrow \{b\})=0.91 > c(\{a\} \rightarrow \{b\})=0.67$$  
and  
$$c(\{\neg a\} \rightarrow \{b\})=0.91 > s(b)=0.8$$

### Answer to question 5.3.2
1. $$ Odds ratio = \frac{f_{11} \cdot f_{00}}{f_{10} \cdot f_{01}}$$  
**Table 2**
$$\frac{105*62}{87*40}=\frac{6510}{3480}=1.87$$  
**Table 3**  
*College students*
$$\frac{2*20}{9*5}=\frac{40}{45}=0.89$$
*Working adults*
$$\frac{103*42}{78*35}=\frac{4326}{2730}=1.58$$  
When the data is pooled together the direction of the association is positive ($1.87>1$).  
However, when the data is seperated into two groups, the College students group has a negative association ($0.89<1$), and the Working adults group retains the positive association ($1.58>1$).

2. $$\phi-coefficient=\frac{P(X,Y)-P(X)P(Y)}{\sqrt{P(X)(1-P(X))P(Y)(1-P(Y)}}$$  
**Table 2**
$$\frac{\frac{105}{294}-\frac{192}{294}\frac{145}{294}}{\sqrt{\frac{192}{294}(1-\frac{192}{294})\frac{145}{294}(1-\frac{145}{294})}}=0.15$$
**Table 3**  
*College students*
$$\frac{\frac{2}{36}-\frac{11}{36}\frac{7}{36}}{\sqrt{\frac{11}{36}(1-\frac{11}{36})\frac{7}{36}(1-\frac{7}{36})}}=-0.02$$
*Working adults*
$$\frac{\frac{103}{258}-\frac{181}{258}\frac{138}{258}}{\sqrt{\frac{181}{258}(1-\frac{181}{258})\frac{138}{258}(1-\frac{138}{258})}}=0.11$$  
When the data is pooled together the direction of the association is positive ($0.15>0$).  
However, when the data is seperated into two groups, the College students group has a negative association ($-0.02<0$), and the Working adults group retains the positive association ($0.11>0$).

3. $$Interest = \frac{P(A,B)}{P(A) \cdot P(B)}$$  
**Table 2**
$$\frac{\frac{105}{294}}{\frac{192}{294} \cdot \frac{145}{294}}=1.11$$
**Table 3**  
*College students*
$$\frac{\frac{2}{36}}{\frac{11}{36} \cdot \frac{7}{36}}=0.94$$
*Working adults*
$$\frac{\frac{103}{258}}{\frac{181}{258} \cdot \frac{138}{258}}=1.06$$  
When the data is pooled together the direction of the association is positive ($1.11>1$).  
However, when the data is seperated into two groups, the College students group has a negative association ($0.94<1$), and the Working adults group retains the positive association ($1.06>1$).
