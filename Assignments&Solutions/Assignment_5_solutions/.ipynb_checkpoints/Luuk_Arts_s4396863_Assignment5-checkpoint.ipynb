#### 5.2.4
(0.5 points) Often we are interested in rules with high confidence. Is it possible for
itemsets to have very low support but still have a very high confidence?

###### Answer 5.2.4:
Yes.

For an association rule of the form X -> Y, where X and Y are itemsets, support measures the fraction of transactions that contain both X and Y and confidence measures how often items in Y appear in transactions that contain X. 

So, for example, if two items are always bought together, but they are only bought in a very small percentage of the transactions, then the support for the itemset containing those two items will be very low, but the confidence will be very high.

 Suppose we have market basket data consisting of 100 transactions and 20 items. $$
c(  \text{ {~a } } \rightarrow   \text{{b }})> c(  \text{ {a } } \rightarrow   \text{{b }})
$$
and
$$
c(  \text{ {~a } } \rightarrow   \text{{b }})> s( {b })
$$

where $ \text{c(.)}$  denotes the rule confidence and  $ \text{s(.)}$ denote the support of an itemset. 

Hint: To prove the statement rewrite the confidence and the support of the rule in terms of probabilities: 
$$
c(  \text{ {a } } \rightarrow   \text{{b }})=  \frac{P( {a,b })}{P( {a })}
$$

and

$$
s( {b })=P( {b })
$$

Further more, make use of probability rules such as $p(b) = p(\text{~}a,b) + p(a,b)$ and $p(\text{~}a) = 1 - p(a)$ and algebraic rules such as $$ \frac{p(a,b)}{p(b)} < p(a) \rightarrow 1 - \frac{p(a,b)}{p(b)} > 1 - p(a) $$


###### Answer 5.3.1:
1) Confidence = 0.3 / 0.45 = 0.66666666666. So the rule is interesting according to the confidence measure, because it passes the confidence treshold of 60%.

2) Lift = 0.66666666666 / 0.8 = 0.8333, which is smaller than 1, so the relationship between item a and item b is that they're negatively associated. 

3) In the results of parts (1) and (2) we can see that despite the fact that the rule passes the confidence treshold it is still uninteresting, because the confidence for item b is already 0.8 which is higher than the confidence of the association rule, meaning that if someone buys item a it actually decreases the probability of them buying b.

So based on the results we can conclude that a rule that's interesting according to the confidence measure can still be uninteresting.

4) 
Rewritten confidence and support of the rules:

$
c(  \text{ {a } } \rightarrow   \text{{b }})=  \frac{P( {a,b })}{P( {a })}
$

$
c(  \text{ {~a } } \rightarrow   \text{{b }})=  \frac{P( {\tilde{}a,b })}{P( {\tilde{}a })}
$

$
s( {b })=P( {b })
$

To prove:

If
$
\frac{P( {a,b })}{P( {a })} < P(b)
$,
 
then
$
\frac{P(\tilde{}a,b)}{P(\tilde{}a)} > P(b)
$

and
$
\frac{P(\tilde{}a,b)}{P(\tilde{}a)} > \frac{P(a,b)}{P(a)}
$

Proof:

Starting off with proving the first equation, we first rewrite the equation:

$
\frac{P(\tilde{}a,b)}{P(\tilde{}a)} > P(b)
$

$
\frac{P(b) - P(a,b)}{1 - P(a)} > P(b)
$

$
\frac{P(b) - P(a,b)}{1 - P(a)} > \frac{P(b) - P(a)P(b)}{1 - P(a)}
$

$ P(b) - P(a,b) > P(b) - P(a)P(b) $

$ -P(a,b) > - P(a)P(b) $

$ P(a,b) < P(a)P(b) $

Now to prove this we use the given assumption $\frac{P(a,b)}{P(a)} < P(b)$, which can be rewritten as $\frac{P(a,b)}{P(a)} < \frac{P(a)P(b)}{P(a)}$, from which follows that $P(a,b) < P(a)P(b)$, completing the proof.

Now to prove the second equation:

$
\frac{P(\tilde{}a,b)}{P(\tilde{}a)} > \frac{P(a,b)}{P(a)}
$

$
\frac{P(b) - P(a,b)}{1 - P(a)} > \frac{P(a,b)}{P(a)}
$

$
\frac{P(a)P(b) - P(a)P(a,b)}{P(a) - P(a)P(a)} > \frac{P(a,b) - P(a)P(a,b)}{P(a) - P(a)P(a)}
$

$
P(a)P(b) - P(a)P(a,b) > P(a,b) - P(a)P(a,b)
$

$ P(a)P(b) > P(a,b)$

Since the alternative notation $P(a,b) < P(a)P(b)$ was already proven in the proof for the first equation, the whole proof is now complete.
