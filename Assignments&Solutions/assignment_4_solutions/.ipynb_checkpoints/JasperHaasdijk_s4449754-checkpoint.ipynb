## 4.1 *k*-means clustering
In this part of the exercise we will investigate *k*-means clustering. In *k*-means each of the data points are assigned to the cluster in closest proximity according to some measure of distance between cluster centers and data points. When the distance is given by the squared euclidian distance, the centers are also called centroids. Once the data points have been assigned, each cluster center is updated to be placed at the center of the data points that are assigned to the cluster. This continues iteratively, usually until the assignment of data points to centers no longer changes or until a maximal number of iterations is reached.

####  4.1.1b - Figure
* The axis labels have been named after the requirements of this exercise, namely `classification error` in percentage, and `Max tree depth` as values of `K`. In the title of each subplot is clearly specified which data set it represents.

**Description**  
In the figures plotted by the script below, you can see the synth data sets and their associated validity measurements.

####  4.1.1b - Script

````python
import sys
sys.path.append('Toolbox')
import clusterVal

from scipy.io import loadmat
import sklearn.cluster as skc
import matplotlib.pyplot as plt
import numpy as np

filenames = ["Data/synth1.mat", "Data/synth2.mat", "Data/synth3.mat", "Data/synth4.mat"]

# for keeping track of [synth{}][K value] validity measures
entropy, purity = np.empty((4, 10), dtype=object), np.empty((4, 10), dtype=object)
rand, jaccard = np.empty((4, 10), dtype=object), np.empty((4, 10), dtype=object)

for i, filename in enumerate(filenames):
    synth_data = loadmat(filename, struct_as_record=False, squeeze_me=True)

    X, y = synth_data['X'], synth_data['y']  # matrix X, class vector y

    for K in range(1, 11):
        centroids, clusters, inertia = skc.k_means(X, K)
        entropy[i, K-1], purity[i, K-1], rand[i, K-1], jaccard[i, K-1] = clusterVal.clusterVal(y, clusters)

xdata = list(range(1, 11))
plt.figure(figsize=(12, 20))
hfont = {'fontname': 'serif'}

for plot in range(0, 4):
    plt.subplot("41{}".format(plot+1))
    plt.plot(xdata, entropy[plot], '-b', label='Entropy')
    plt.plot(xdata, purity[plot], '-r', label='Purity')
    plt.plot(xdata, rand[plot], '-y', label='Rand statistic')
    plt.plot(xdata, jaccard[plot], '-g', label='Jaccard coefficient')
    plt.legend(loc='upper left', prop={'size': 9})
    plt.xlabel("Max tree depth (K)", **hfont)
    plt.ylabel("Classification error (%)", **hfont)
    plt.title("Synth{}: Plotting the average of the training and test classification error as a function of the tree depth"
              .format(plot+1), **hfont)

plt.tight_layout()
plt.show()
````

> 4.1.2b (1 point) How can the cluster validity measures be used to select the best number of clusters?
What happens when more than four clusters are used to model the data?

####  4.1.2b - Answer
The cluster validity measures can be used to select the best number of clusters by looking at the classification errors. The lower the error rate, the better the cluster validity, the better the choosen number of clusters. 

Of course some measures have their quirks that need to be taken into account. For example when using the `purity` measure we can trick the error rate by using a huge amount of clusters, since this measure doesn't penalize the usage of many clusters. When using the `rand statistic` measure the `FP` and `FN` values in the computation are equally weighed. This might not be desirable in your particular dataset. I cannot say anything about the _synth_ datasets with regard to these quirks, as there has been given no context with the data.

Since there are only 4 different classes in the data set, after more than four clusters are used to model the data, multiple clusters contain data points from the same class.

For fun I have added a total classification error data plot. This figure shows the total classification error of the data sets as a function of the tree depth (`K`). This shows a combined score of how well the data clustering performed.

````python
xdata = list(range(1, 11))
plt.figure(figsize=(14, 8))
hfont = {'fontname': 'serif'}

total_error = np.empty((4, 10), dtype=object)

for i in range(0, 4):
    total_error[i] = entropy[i] + purity[i] + rand[i] + jaccard[i]
    plt.plot(xdata, total_error[i], label='Synth{}'.format(i+1))

plt.legend(loc='upper left', prop={'size': 9})
plt.xlabel("Max tree depth", **hfont)
plt.ylabel("Total classification error (%)", **hfont)
plt.title("Plotting the total classification error of the data sets as a function of the tree depth", **hfont)
plt.show()
````

The function *dendrogram* creates a plot of the generated tree. The function *fcluster* extracts the cluster from a linkage matrix w.r.t. a given criterion. + You can e.g. type *Z = linkage(X, method='single', metric='euclidean')* to use single linkage with the Euclidean distance measure.
+ To compute a clustering, you can use the function *fcluster()*. For example, type *cls = fcluster(Z, criterion='maxclust', t=4)* to get a maximum of 4 clusters. Type *help(fcluster)* (or read the online documentation) to learn more about what this function does.

#### 4.2.1a - Figure

In the figures below you can see the plotted dendrogram and scatterplot of the clustering of the _synth1_ data set. The colors are there to visualize the different classes and the clusters they belong to.

#### 4.2.1a - Script

````python
from scipy.cluster.hierarchy import *
from scipy.io import loadmat
import clusterPlot

filename = "Data/synth1.mat"
synth_data = loadmat(filename, struct_as_record=False, squeeze_me=True)
X = synth_data['X']
y = synth_data['y']

plt.figure(figsize=(15, 20))

plt.subplot(211)
plt.title("Plotting a dendrogram of the clusters")
Z = linkage(X, method='single', metric='euclidean')
plot1 = dendrogram(Z, p=4, truncate_mode='lastp')


plt.subplot(212)
cls = fcluster(Z, criterion='maxclust', t=4)
clusterPlot.clusterPlot(X, cls, y=y)

plt.title("Plotting a scatterplot of the clusters")
plt.xlabel("Attribute A")
plt.ylabel("Attribute B")
plt.show()
````

> 4.2.1b (1 point) Change the linkage methods (single, complete, average) and explain how this changes the dendrogram.
Then plot a dendrogram and a scatterplot of the clusters for the *synth2*, *synth3*, and *synth4* data sets..

#### 4.2.1b - Answer
* Explain how changing the linkage method (single, complete, average) changes the dendrogram
    * Single; starts grouping the closest points in the different clusters. This means that the lower you search in the dendrogram the closer the nodes actually are from one another
    * Complete; starts grouping the least similar points in the different clusters. This means that the lower you search in the dendrogram the farther the nodes actually are from one another
    * Average; as the word suggests, starts grouping the points in the different clusters based on the average point proximity. This means that the lower you search in the dendrogram the more 'average' the nodes' proximity is.
