## 4.1 *k*-means clustering
In this part of the exercise we will investigate *k*-means clustering. In *k*-means each of the data points are assigned to the cluster in closest proximity according to some measure of distance between cluster centers and data points. When the distance is given by the squared euclidian distance, the centers are also called centroids. Once the data points have been assigned, each cluster center is updated to be placed at the center of the data points that are assigned to the cluster. This continues iteratively, usually until the assignment of data points to centers no longer changes or until a maximal number of iterations is reached.

> 4.1.2a (1 point) Repeat Exercise 4.1.1, but this time perform *k*-means clustering for *K = 1, ..., 10* clusters. For each value of *K* compute the four cluster validity measures mentioned above. Plot the cluster validity measures as a function of *K*
 + *Use the clusterVal() function to compute the cluster validity. Don't forget to import it from the Toolbox folder. If you are unsure about how to use the function, open the .py file in a text editor and take a look at documentation.*

```python
synth_data1 = loadmat('Data/synth1')
synth_data2 = loadmat('Data/synth2')
synth_data3 = loadmat('Data/synth3')
synth_data4 = loadmat('Data/synth4')

Xlist = []
Xlist.append(synth_data1['X'])
Xlist.append(synth_data2['X'])
Xlist.append(synth_data3['X']) 
Xlist.append(synth_data4['X'])

ylist = []
ylist.append(synth_data1['y'])
ylist.append(synth_data2['y'])
ylist.append(synth_data3['y'])
ylist.append(synth_data4['y'])

for j in range (0,4):
        X = Xlist[j]
        y = ylist[j]

        #Calculate different cluster validity measures
        clusterValList = []
        for k in range(1, 11):
            centroid, label, inertia = cluster.k_means(X, k)
            clusterValList.append(clusterVal(y, label))

        #Retrieve different cluster validity measures
        entropy = [x[0] for x in clusterValList]
        purity = [x[1] for x in clusterValList]
        rand = [x[2] for x in clusterValList]
        jaccard = [x[3] for x in clusterValList]

        #Plot the results
        K = range(1,11)
        plt.xlabel('Maximum Tree Depth')
        plt.ylabel('Classification Error')
        plt.title('Classification Errors for the test and training data')

        plt.plot(K, entropy, K, purity, K, rand, K, jaccard)
        plt.legend(['Entropy', 'Purity', 'Rand', 'Jaccard'], loc='upper left')
        plt.xlabel('K')
        plt.ylabel('Similarity Measure Value')
        plt.title('synth%s , Different similarity measures plotted for different Ks' %(j+1))
        plt.show()
```

> 4.1.2b (1 point) How can the cluster validity measures be used to select the best number of clusters?
What happens when more than four clusters are used to model the data?

Cluster validity measures can be used to select the best number of clusters. This is done by measuring the degree to which predicted class labels correspond to the actual class labels. So when this validity measure is better, the number of clusters chosen is also better. 

When more than four clusters are used to model the data:

**Purity**:

When the number of clusters goes up, the purity will go up. Purity shows to which extent a cluster contains objects of a single class. When the number of clusters goes up, the different clusters will seperated better, thus there will be less different classes in each cluster. This means the purity will go up. The real classes will be split up into more clusters. For example, when the number of clusters is the number of datapoints, the purity will be 1, because each cluster contains only one class: the class of the one datapoint that it contians.

**Entropy**: 

When the number of clusters goes up, the number of different classes in a cluster will go down. This means that the probability that a member of cluster j belongs to class i (pij) will go towards 1 for one class, and the probability will go towards 0 for all the other classes. When pij goes up, the log of that number will be a negative number that gets closer to 0. The sum of these numbers will be smaller, thus the sum * -1 will be closer to zero. The entropy will thus go down when the number of clusters goes up.

**Rand statistic & Jaccard Coefficient**:

Rand statistic = (f00 + f11)/(f00 + f01 + f10 + f11) and the Jaccard Coefficient = (f11)/(f00 + f01 + f10 + f11). f11 is the number of points that have the same class that also have the same cluster. f10 is the number of points that have the same class but that are in a different cluster. f01 is the number of points that have a different class but are in the same cluster. f00 is the number of datapoints that are in a different class and are in a different cluster.

*Rand Statistic*: When the number of clusters goes up, the Rand statistic will go down, because f11 will decrease and f10 will increase. The classes will be split up into different clusters. 

*Jaccard Coefficient*: When the number of clusters goes up, the Jaccard coefficient will go down, for the same reason as the Rand statistic. 

The Jaccard coefficient decreses more, because for this measure, f00 is not used. f00 stays the same when the number of clusters is increased. For the Rand Statistic, f11 and f00 are added together. When f00 is added the decrease in number of f11 data points has a smaller effect.

*k*-means clustering has many different applications, one of which is data compression. A data set can be compressed by performing *k*-means clustering and then representing each data object by its cluster center. Thus, the only data that need to be stored are the *K* cluster centers and the *N* cluster indices.

## 4.2 Hierarchical Clustering

In this part of the exercie we will consider hierarchical clustering based on the functions from the package *scipy.cluster.hierarchy*. The function *linkage()* forms a sample to sample distance matrix according to a given distance metric, and creates the linkages between data points forming the hierarchical cluster tree. The function *dendrogram* creates a plot of the generated tree. The function *fcluster* extracts the cluster from a linkage matrix w.r.t. a given criterion. Use *help* for the three functions (or read the online documentation) and inspect what distance metrics and linkage functions are implemented.

> 4.2.1a (1 point) Load the data set from *Data/synth1*. Cluster the data using hierarchical clustering with single linkage using the Euclidean distance measure. Cluster the data into 4 clusters by cutting off the dendrogram at a threshold. Plot a dendrogram and a scatter plot of the clusters.
+ The function *linkage()* computes the hierarchical clustering, resulting in a matrix representing the hierarchy of clusterings. Type help(linkage) (or read the online documentation) to learn how to use it.
+ You can e.g. type *Z = linkage(X, method='single', metric='euclidean')* to use single linkage with the Euclidean distance measure.
+ To compute a clustering, you can use the function*fcluster()*. For example, type *cls = fcluster(Z, criterion='maxclust', t=4)* to get a maximum of 4 clusters. Type *help(fcluster)* (or read the online documentation) to learn more about what this function does.
+ To plot a dendrogram, you can use the *dendrogram()* function
+ Again, you can use the function *clusterplot()* to plot a scatter plot of the clustering.

```python
import scipy.cluster.hierarchy as sch

synth_data = loadmat('Data/synth1')
X = synth_data['X']
y = synth_data['y']

Z = sch.linkage(X, method = 'single', metric = 'euclidean')
cls = sch.fcluster(Z, criterion = 'maxclust', t = 4)
dendro = sch.dendrogram(Z)
plt.title('Dendrogram synth1')
plt.ylabel('Euclidian distance')
plt.xlabel('Datapoints')
plt.figtext(.95, .7, 'synth1 data clustered using\nhierarchical clustering,\nusing Euclidean distance as\ndistance measure, and \n"single" as method', rotation = 'horizontal')
plt.show()

clusterPlot(X, cls)
plt.title('Scatterplot synth1')
plt.xlabel('x')
plt.ylabel('y')
plt.figtext(.95, .7, 'synth1 data clustered using\nhierarchical clustering,\nusing Euclidean distance as\ndistance measure, and \n"single" as method', rotation = 'horizontal')
plt.show()
```

> 4.2.1b (1 point) Change the linkage methods (single, complete, average) and explain how this changes the dendrogram.
Then plot a dendrogram and a scatterplot of the clusters for the *synth2*, *synth3*, and *synth4* data sets..

```python
synth_data1 = loadmat('Data/synth1')
X = []
X1 = synth_data1['X']
X.append(X1)
synth_data2 = loadmat('Data/synth2')
X2 = synth_data2['X']
X.append(X2)
synth_data3 = loadmat('Data/synth3')
X3 = synth_data3['X']
X.append(X3)
synth_data4 = loadmat('Data/synth4')
X4 = synth_data4['X']
X.append(X4)

for i in range(0,4):
    plt.figure(figsize = (15,10))
    
    # Create the dendrograms
    plt.subplot(2,3,1)
    cls = sch.fcluster(sch.linkage(X[i], method='single', metric='euclidean'), criterion='maxclust', t=4)
    dendrogram = sch.dendrogram(sch.linkage(X[i], method='single', metric='euclidean'))
    plt.title('Dendrogram of synth%s, method = "single"' %(i+1))
    plt.xlabel('Datapoints')
    plt.ylabel('Euclidian distance')
    plt.subplot(2,3,2)
    cls = sch.fcluster(sch.linkage(X[i], method='complete', metric='euclidean'), criterion='maxclust', t=4)
    dendrogram = sch.dendrogram(sch.linkage(X[i], method='complete', metric='euclidean'))
    plt.title('Dendrogram of synth%s, method = "complete"' %(i+1))
    plt.xlabel('Datapoints')
    plt.ylabel('Euclidian distance')
    plt.subplot(2,3,3)
    cls = sch.fcluster(sch.linkage(X[i], method='average', metric='euclidean'), criterion='maxclust', t=4)
    dendrogram = sch.dendrogram(sch.linkage(X[i], method='average', metric='euclidean'))
    plt.title('Dendrogram of synth%s, method = "average"' %(1+i))
    plt.xlabel('Datapoints')
    plt.ylabel('Euclidian distance')
    
    # Create the scatterplots
    plt.subplot(2,3,4)
    clusterplot = clusterPlot(X[i], cls)
    plt.title('Scatterplot of synth%s, method = "single"' %(i+1))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.subplot(2,3,5)
    clusterplot = clusterPlot(X[i], cls)
    plt.title('Scatterplot of synth%s, method = "complete"' %(i+1))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.subplot(2,3,6)
    clusterplot = clusterPlot(X[i], cls)
    plt.title('Scatterplot of synth%s, method = "average"' %(i+1))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

Different methods are different ways of comparing the distance between the clusters. 

For method='single', a datapoint will be put together with the closest data point. This often means that a datapoint will be added to an existing big cluster, because one of the points in that cluster is close to the data point.

For method = 'complete', small groups of data points will form different small clusters first, before those clusters will be added together. This is because a point will only be added to an existing bigger cluster if it is closer to the furthest data point in that cluster than to all other data points that are not in that cluster.

Method = 'average' is inbetween these two methods, thus sometimes data points will form a new cluster, and sometimes data points will be added to existing bigger clusters.
