```python
import seaborn as sn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_confusion_matrix(cm,y):
    df_cm = pd.DataFrame(cm, index = [i+1 for i in np.unique(y)],
                  columns = [i+1 for i in np.unique(y)])
    plt.figure()
    sn.heatmap(df_cm, annot=True)
    plt.title('Confusion matrix')
    plt.xlabel('Predicted class')
    plt.ylabel('Actual class')
    plt.show()
```

```python

from scipy.io import loadmat
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

## Synth 1
mat_data = loadmat('Data/synth1.mat')
X = mat_data['X']
y = mat_data['y']

for c in range(4):
    plt.scatter(X[(y==c).flatten(),0],X[(y==c).flatten(),1],label='{}'.format(c+1))
plt.title('Synth1')
plt.legend()
plt.show()

K=1
dist=2 #p = 1 is city block/manhattan, p = 2 is euclidean
knclassifier = KNeighborsClassifier(n_neighbors=K, p=dist);
knclassifier.fit(X, y.ravel())
y_hat = knclassifier.predict(X)


for c in range(4):
    plt.scatter(X[(y_hat==c).flatten(),0],X[(y_hat==c).flatten(),1])
plt.title('Synth1 classification')
plt.show()

cm = confusion_matrix(y.ravel(), y_hat);
accuracy = 100.0*float(cm.diagonal().sum())/cm.sum(); 
error_rate = 100-accuracy;
plot_confusion_matrix(cm, y)
print('Accuracy: {}, Error Rate: {}'.format(accuracy, error_rate))

## Synth 2
mat_data = loadmat('Data/synth2.mat')
X = mat_data['X']
y = mat_data['y']

for c in range(4):
    plt.scatter(X[(y==c).flatten(),0],X[(y==c).flatten(),1],label='{}'.format(c+1))
plt.title('Synth2')
plt.legend()
plt.show()

K=1
dist=2 #p = 1 is city block/manhattan, p = 2 is euclidean
knclassifier = KNeighborsClassifier(n_neighbors=K, p=dist);
knclassifier.fit(X, y.ravel())
y_hat = knclassifier.predict(X)


for c in range(4):
    plt.scatter(X[(y_hat==c).flatten(),0],X[(y_hat==c).flatten(),1])
plt.title('Synth2 classification')
plt.show()

cm = confusion_matrix(y.ravel(), y_hat);
accuracy = 100.0*float(cm.diagonal().sum())/cm.sum();
error_rate = 100-accuracy;
plot_confusion_matrix(cm, y)
print('Accuracy: {}, Error Rate: {}'.format(accuracy, error_rate))



## Synth 3
mat_data = loadmat('Data/synth3.mat')
X = mat_data['X']
y = mat_data['y']

for c in range(4):
    plt.scatter(X[(y==c).flatten(),0],X[(y==c).flatten(),1],label='{}'.format(c+1))
plt.title('Synth3')
plt.legend()
plt.show()

K=1
dist=2 #p = 1 is city block/manhattan, p = 2 is euclidean
knclassifier = KNeighborsClassifier(n_neighbors=K, p=dist);
knclassifier.fit(X, y.ravel())
y_hat = knclassifier.predict(X)


for c in range(4):
    plt.scatter(X[(y_hat==c).flatten(),0],X[(y_hat==c).flatten(),1])
plt.title('Synth3 classification')
plt.show()

cm = confusion_matrix(y.ravel(), y_hat);
accuracy = 100.0*float(cm.diagonal().sum())/cm.sum();
error_rate = 100-accuracy;
plot_confusion_matrix(cm, y)
print('Accuracy: {}, Error Rate: {}'.format(accuracy, error_rate))



## Synth 4
mat_data = loadmat('Data/synth4.mat')
X = mat_data['X']
y = mat_data['y']

for c in range(4):
    plt.scatter(X[(y==c).flatten(),0],X[(y==c).flatten(),1],label='{}'.format(c+1))
plt.title('Synth4')
plt.legend()
plt.show()

K=1
dist=2 #p = 1 is city block/manhattan, p = 2 is euclidean
knclassifier = KNeighborsClassifier(n_neighbors=K, p=dist);
knclassifier.fit(X, y.ravel())
y_hat = knclassifier.predict(X)


for c in range(4):
    plt.scatter(X[(y_hat==c).flatten(),0],X[(y_hat==c).flatten(),1])
plt.title('Synth4 classification')
plt.show()

cm = confusion_matrix(y.ravel(), y_hat);
accuracy = 100.0*float(cm.diagonal().sum())/cm.sum(); 
error_rate = 100.0-accuracy;
plot_confusion_matrix(cm, y)
print('Accuracy: {}, Error Rate: {}'.format(accuracy, error_rate))


```

```python
from sklearn import model_selection
import pandas as pd
import numpy as np

data = pd.read_excel('Data/iris.xls')
print(data.head())
X = data['Sepal Length  Sepal Width  Petal Length  Petal Width'.split('  ')].values
y = data['Type'].values

N = len(y)
print(N)
L=40 # max. number of neighbors
CV = model_selection.LeaveOneOut()
errors = np.zeros((N,L))
i=0
for train_index, test_index in CV.split(X):
    X_train = X[train_index,:]
    y_train = y[train_index]
    X_test = X[test_index,:]
    y_test = y[test_index]

    for l in range(1,L+1):
        knclassifier = KNeighborsClassifier(n_neighbors=l);
        knclassifier.fit(X_train, y_train.ravel());
        y_est = knclassifier.predict(X_test);
        errors[i,l-1] = np.sum(y_est[0]!=y_test[0])

    i+=1
    
plt.plot(np.mean(errors,axis=0))
plt.xlabel('Number of neighbors')
plt.ylabel('Classification error rate (%)')
plt.show()
```

```python

cv = model_selection.KFold(n_splits=10)
train_error = np.zeros(10)
test_error = np.zeros(10)

k=0
for train_index,test_index in cv.split(X):
    X_train = X[train_index,:]
    y_train = y[train_index]
    X_test = X[test_index,:]
    y_test = y[test_index]
    print('Fold {}'.format(k))
   
    best_train = np.inf
    best_test = np.inf
    for i in range(5):
        perceptron = mlp.MLP(X_train,y_train)
        w1,w2,c = perceptron.learn_weights(verbose=False)
        y_hat = perceptron.forwardprop(w1,w2,X=X_test)
        if(c < best_train):
            best_train=c
        e_test = 1-np.mean(y_hat == y_test)
        if(e_test < best_test):
            best_test=e_test
    train_error[k] = best_train
    test_error[k] = best_test
    k+=1

print('Average training error: {} \nAverage test error: {}'.format(np.mean(train_error),np.mean(test_error)))

perceptron = mlp.MLP(X,y)
w1,w2,c = perceptron.learn_weights()
perceptron.plot_boundaries(w1,w2,X)
```
